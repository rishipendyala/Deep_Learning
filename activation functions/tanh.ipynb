## 🧠 Tanh Activation Function

The **Tanh (hyperbolic tangent) function** maps any real-valued input into **-1 to 1**.  
It is defined as:

$$
\tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
$$

**Properties:**

- **Range:** -1 to 1  
- **Shape:** S-shaped curve (sigmoid-like)  
- **Symmetry:** Centered at 0 (unlike sigmoid)  
- **Output at 0:** 0  

**Intuition:**

- Positive inputs → output → 1  
- Negative inputs → output → -1  
- Centered around 0 → helps with gradient flow  

---

### ✅ Use Cases

- Hidden layers in neural networks (especially before ReLU)  
- Sequence models (RNNs, LSTMs)  
- Tasks where zero-centered output is useful  

---

### 🌟 Advantages

- **Non-linear:** Allows the network to learn complex patterns  
- **Zero-centered output:** Improves gradient-based optimization  
- **Better gradient behavior than sigmoid:** Helps mitigate vanishing gradient (partially)  
- **Differentiable:** Suitable for backpropagation  

---

### ⚠️ Disadvantages

- **Vanishing gradient:** For large |x|, gradients approach 0 → slower learning in deep networks  
- **Sensitive to outliers:** Saturated regions reduce learning efficiency  
- **Not always best for deep networks:** ReLU variants often preferred  

---

### 📈 Derivative

The derivative is used in backpropagation:

$$
\frac{d}{dx} \tanh(x) = 1 - \tanh^2(x)
$$

- Maximum gradient at x = 0  
- Gradient decreases for large positive or negative inputs  

---

### 💡 When to Use

- Data normalized or centered around 0  
- Shallow networks (few layers)  
- Tasks where negative values carry meaningful information  

> For deeper networks, consider ReLU or Leaky ReLU to avoid vanishing gradients.


```python
# Plot tanh and its derivative
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5, 5, 100)
tanh_x = np.tanh(x)
dtanh_x = 1 - tanh_x**2
plt.plot(x, tanh_x, label='tanh(x)')
plt.plot(x, dtanh_x, label="tanh'(x)")
plt.title('Tanh Activation Function and its Derivative')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

```python
import torch
import torch.nn as nn
class SimpleNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(SimpleNN, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.tanh = nn.Tanh()
        self.fc2 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        out = self.fc1(x)
        out = self.tanh(out)
        out = self.fc2(out)
        return out
```
